### Aqui vamos baixar os dados da climatologia do MERGE e do SAMET... Isso já é disponibilizado na página do CPTEC/INPE

Neste exemplo optei por usar a linguagem de script Bash para fazer um loop de datas, substituindo essa variável no link para download com o comando *wget*


```python
%%bash
#  Aqui vamos fazer um loop de data em bash
# e usar o comando "wget" para download dos links
dataini='20250101'
for ((d=0; d<=365; d++))
do
  data=$(date -d"$dataini +$d days" +%d%b | tr '[:upper:]' '[:lower:]')
  echo $data
  wget https://ftp.cptec.inpe.br/modelos/tempo/SAMeT/CLIMATOLOGY/DAILY_AVERAGE/TMAX/SAMeT_CPTEC_TMAX_00Z$data.nc -O climatology/SAMeT_CPTEC_TMAX_00Z$data.nc
  wget https://ftp.cptec.inpe.br/modelos/tempo/SAMeT/CLIMATOLOGY/DAILY_AVERAGE/TMIN/SAMeT_CPTEC_TMIN_00Z$data.nc -O climatology/SAMeT_CPTEC_TMIN_00Z$data.nc
  wget https://ftp.cptec.inpe.br/modelos/tempo/MERGE/GPM/CLIMATOLOGY/DAILY_AVERAGE/MERGE_CPTEC_12Z$data.nc -O climatology/MERGE_CPTEC_12Z$dsata.nc
done
```

### Aqui, apenas para fins didáticos, vamos fazer o mesmo loop, mas para download dos dados de 2023 e 2024, com *Python*

```python
import datetime as dt
import requests
import os

# Define as datas
dataini = dt.datetime(2023, 1, 1)
datafim = dt.datetime(2024, 12, 31)

# Loop pelas datas
data = dataini
while data <= datafim:
    ano = data.strftime('%Y')
    mes = data.strftime('%m')
    dia = data.strftime('%d')
    data_str = f'{ano}{mes}{dia}'
    print(f'Iniciando download de {ano}-{mes}-{dia}')

    # MERGE
    resposta = requests.get(f'https://ftp.cptec.inpe.br/modelos/tempo/MERGE/GPM/DAILY/{ano}/{mes}/MERGE_CPTEC_{data_str}.grib2')
    with open(f'prec/MERGE_CPTEC_{data_str}.grib2', 'wb') as f:
        f.write(resposta.content)

    # TMAX
    resposta = requests.get(f'https://ftp.cptec.inpe.br/modelos/tempo/SAMeT/DAILY/TMAX/{ano}/{mes}/SAMeT_CPTEC_TMAX_{data_str}.nc')
    with open(f'tmax/SAMeT_CPTEC_TMAX_{data_str}.nc', 'wb') as f:
        f.write(resposta.content)

    # TMIN
    resposta = requests.get(f'https://ftp.cptec.inpe.br/modelos/tempo/SAMeT/DAILY/TMIN/{ano}/{mes}/SAMeT_CPTEC_TMIN_{data_str}.nc')
    with open(f'tmin/SAMeT_CPTEC_TMIN_{data_str}.nc', 'wb') as f:
        f.write(resposta.content)

    data += dt.timedelta(days=1)

print('Downloads finalizados!')
```
