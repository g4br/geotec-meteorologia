# **CENTROS NACIONAIS**

Aqui está a tabela com os principais sistemas e órgãos **federais** de monitoramento de tempo, clima e recursos hídricos no Brasil:

## Sistemas e Órgãos Nacionais

| Instituição / Sistema | Foco Principal | Vinculação | Site |
|---|---|---|---|
| **INMET** – Instituto Nacional de Meteorologia | Monitoramento, previsão do tempo e clima, avisos severos, rede de estações | Ministério da Agricultura e Pecuária | [portal.inmet.gov.br](https://portal.inmet.gov.br) |
| **CPTEC/INPE** – Centro de Previsão de Tempo e Estudos Climáticos | Previsão numérica, monitoramento climático, subsazonal e sazonal, satélite, qualidade do ar | MCTI / INPE | [cptec.inpe.br](https://www.cptec.inpe.br) |
| **CEMADEN** – Centro Nacional de Monitoramento e Alertas de Desastres Naturais | Alertas de desastres naturais (chuvas, deslizamentos, secas), sistema SALVAR | MCTI | [cemaden.gov.br](https://www.gov.br/cemaden/pt-br) |
| **ANA** – Agência Nacional de Águas e Saneamento Básico | Gestão e coordenação da Rede Hidrometeorológica Nacional (RHN) | MMA | [ana.gov.br](https://www.gov.br/ana/pt-br) |
| **HidroWeb** – Sistema de Informações Hidrológicas | Dados hidrometeorológicos históricos e em tempo real (pluviometria, fluviometria) | ANA | [snirh.gov.br/hidroweb](https://www.snirh.gov.br/hidroweb) |
| **SNIRH** – Sistema Nacional de Informações sobre Recursos Hídricos | Portal integrador de dados hídricos nacionais | ANA | [snirh.gov.br](https://www.snirh.gov.br) |
| **SAR** – Sistema de Acompanhamento de Reservatórios | Monitoramento de volumes e níveis de reservatórios | ANA | [sar.ana.gov.br](https://sar.ana.gov.br) |
| **SGB/CPRM** – Serviço Geológico do Brasil | Rede Hidrometeorológica Nacional de Referência (RHNR), geologia, hidrologia | MME | [sgb.gov.br](https://www.sgb.gov.br) |
| **Monitor de Secas** | Monitoramento e classificação de secas por região | ANA + parceiros estaduais | [monitordesecas.ana.gov.br](https://monitordesecas.ana.gov.br) |
| **CENAD** – Centro Nacional de Gerenciamento de Riscos e Desastres | Coordenação de resposta a desastres, integra alertas do CEMADEN | Mintegration / MDR | [gov.br/defesacivil](https://www.gov.br/defesacivil) |
| **INPE – Satélites e Imagens** | Imagens de satélite, desmatamento (PRODES/DETER), queimadas | MCTI / INPE | [inpe.br](https://www.inpe.br) |
| **Dados Abertos ANA** | Portal de dados abertos hidrológicos e de saneamento | ANA | [dadosabertos.ana.gov.br](https://dadosabertos.ana.gov.br) |

***

O INMET representa o Brasil junto à **Organização Meteorológica Mundial (OMM)** e é responsável pelo tráfego de mensagens meteorológicas da América do Sul. A ANA coordena a RHN com quase **4.800 estações de monitoramento** distribuídas em todo o país, sendo 1.960 fluviométricas e 2.840 pluviométricas. Já o CEMADEN integra dados de múltiplas fontes para emitir alertas de risco de desastres, com seu sistema SALVAR voltado à visualização de áreas críticas. [portal.inmet.gov](https://portal.inmet.gov.br/sobre)
